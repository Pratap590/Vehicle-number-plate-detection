```python
import torch
import easyocr
import cv2

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='best (1).pt')

# Initialize OCR engine
reader = easyocr.Reader(['en','ar'])

# Load video
cap = cv2.VideoCapture('demo.mp4')

# Loop over video frames
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection with YOLOv5
    results = model(frame)

    # Iterate over detected objects
    for bbox in results.pred[0].tolist():
        # Extract text region from bounding box
        x1, y1, x2, y2 = [int(coord) for coord in bbox[:4]]
        text_region = frame[y1:y2, x1:x2, :]

        # Perform OCR on text region
        ocr_result = reader.readtext(text_region)
        if ocr_result:
            text, (left, top, right, bottom) = ocr_result[0]

            # Draw bounding box and text on original frame
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
            cv2.putText(frame, text, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)

    # Display frame with OCR results
    cv2.imshow('YOLOv5 + Easy OCR', frame)
    cv2.waitKey(1)

# Release video capture
cap.release()
cv2.destroyAllWindows()
```
